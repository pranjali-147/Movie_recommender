```python
def clean_list(x):
    if isinstance(x, str):
        return [i.strip() for i in x.split(',')]
    elif isinstance(x, list):
        return x
    else:
        return []
```

```python
df['genre'] = df['genre'].apply(clean_list)
df['actors'] = df['actors'].apply(clean_list)
df['director'] = df['director'].apply(clean_list)
```

```python
df['director'] = df['director'].apply(
    lambda x: [name.replace(" ", "") for name in x]
)

df['actors'] = df['actors'].apply(
    lambda x: [name.replace(" ", "") for name in x]
)
```

```python
df['tags'] = df['overview'] + df['genre'] + df['director'] + df['actors'] 
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x))
```

```python
new_df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
import nltk
```

```python
cv = CountVectorizer(stop_words='english', max_features=5000)
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))

    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
len(cv.get_feature_names_out())
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
def recommend(movie):
    movie_index = new_df[new_df['movie_name'] == movie].index[0]
    distances = similarity[movie_index]

    movies_list = sorted(
        [(i, float(score)) for i, score in enumerate(distances)],
        reverse=True,
        key=lambda x: x[1]
    )[1:31]

    return [new_df.iloc[i[0]].movie_name for i in movies_list]
```

```python
recommend('#Mute')
```

```python
def search_actors(actor_name, top_n=30):
    result = df[df['actors'].apply(
        lambda x: actor_name.lower() in ' '.join(x).lower()
    )]
    result = result.sort_values(by='year', ascending=False)
    return result[['movie_name', 'year', 'actors']].head(top_n)
```
